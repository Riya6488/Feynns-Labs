```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

X = data_final.drop(['Stage'],axis=1).values
y = data_final['Stage'].values
```

```python
print(X.shape)
print(y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,shuffle = True)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
X_train
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
y_train
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
y_train =np.array([ x-1 for x in y_train])
y_test = np.array([ x-1 for x in y_test])
```

```python
y_train
```

```python
y_test
```

```python

y_train = tf.one_hot(y_train,depth =4 , axis=0)
y_test = tf.one_hot(y_test,depth = 4,axis =0 )
```

```python
y_train
```

```python
y_test
```

```python

print(y_train.shape)
print(y_test.shape)
```

```python
y_train_size = y_train.shape[1]
y_test_size = y_test.shape[1]
print(y_train_size)
print(y_test_size)
y_train = np.transpose(y_train)
y_test = np.transpose(y_test)
```

```python
print(y_train.shape)
print(y_test.shape)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
import tensorflow as tf

model = Sequential([
    Dense(32, activation='relu', input_shape=(9,)),
    Dense(64, activation='relu'),
    Dense(64, activation='relu'),
    Dense(16, activation='relu'),
    Dense(4,activation = tf.nn.softmax)
])
```

```python
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
prediction = model.fit(X_train, y_train,
              epochs=2000)
```

```python
model.evaluate(X_test,y_test, verbose=2)
```
